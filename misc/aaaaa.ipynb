```python
min_dist = 40

units = ['Kernkraftwerk GÃ¶sgen', 'Leibstadt', 'KKM Produktion', 'Beznau 1', 'Beznau 2']
unit_idx = {k: None for k in units} # initialise dict
for u in units:
    indices = list(data[data[u] > 200].index)
    if indices:
        unit_idx[u] = indices
    

data_len = len(data.index)

best_option = data_len//2
unit_len = {k: (len(unit_idx[k]) if type(unit_idx[k]) is list else data_len+1) for k in units}

while any(x < data_len+1 for x in list(unit_len.values())):
    min_unit = min(unit_len, key=unit_len.get)
    unit_idx[min_unit] = min(unit_idx[min_unit], key=lambda x:abs(x-best_option))
    ### rules come here ###
    for v in unit_idx.values():
        if type(v) is list:
            active_loop = True
            while active_loop:
                vv = min(v, key=lambda x:abs(x-unit_idx[min_unit]))
                if abs(vv - unit_idx[min_unit]) < min_dist and len(v) > 1:
                    v.remove(vv)
                else:
                    active_loop = False

    unit_len = {k: (len(unit_idx[k]) if type(unit_idx[k]) is list else data_len+1) for k in units}

    print(unit_len)

print(unit_idx)
```
